```python
# Dependencies
import numpy as np

# Python SQL toolkit and Object Relational Mapper
import sqlalchemy
from sqlalchemy.ext.automap import automap_base
from sqlalchemy.orm import Session
from sqlalchemy import create_engine, func
import pandas as pd
```

```python
engine = create_engine("sqlite:///Bank_marketing_campaign99.db")
```

```python
# reflect an existing database into a new model
Base = automap_base()
# reflect the tables
Base.prepare(engine, reflect=True)
```

```python
# Create our session (link) from Python to the DB
session = Session(engine)
```

```python
Base.classes.keys()
```

```python
# Save references to each table
information = Base.classes.Information
outcome = Base.classes.Outcome
```

```python
#Join both of the tables by using an INNER JOIN and matching the "ID" column (which is present in both Tables of the DB)

conn = sqlite3.connect("Bank_marketing_campaign99.db")
cursor = conn.cursor()


# Query for INNER JOIN
sql = '''SELECT * from Information INNER JOIN Outcome ON Outcome.ID = Information.ID;'''

# Executing the query
cursor.execute(sql)
  
# Fetching rows from the result table
result = cursor.fetchall()
for row in result:
    print(row)
  
# Closing the connection
conn.close()
```

```python
#Convert the final table into a DataFrame
df = pd.DataFrame(result, columns = ['ID','Age','job','marital','education','defaultnew','balance','housing','loan','day','duration','campaign','pdays','previous','poutcome', 'month_num', 'ID','y'])
df
```

## Random Forest Classifier (trial


```python
import numpy as np
import pandas as pd
from pathlib import Path
from collections import Counter
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import balanced_accuracy_score
```

## Logistic Regression with Standard Scaler


```python
from sklearn.metrics import confusion_matrix, accuracy_score, classification_report
```

```python
from sklearn.model_selection import train_test_split

X=df.drop(columns="y")

y=df["y"]

X_train, X_test, y_train, y_test=train_test_split( X,y, random_state=1, stratify=y)
```

```python
#scale the data
#While not strictly required for tree-based models, 
#it can be a good idea to scale the data, especially when comparing the performances of different models.

from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
X_scaler=scaler.fit(X_train)
X_train_scaled=X_scaler.transform(X_train)
X_test_scaled=X_scaler.transform(X_test)
```

```python
from sklearn.linear_model import LogisticRegression

model=LogisticRegression(solver="lbfgs",max_iter=200,random_state=1)
model.fit(X_train_scaled,y_train)

```

```python

y_pred=model.predict(X_test_scaled)
```

```python

# Calculated the balanced accuracy score

acc_score=accuracy_score(y_test,y_pred)
acc_score
```

```python
#Evaluate the model

cm=confusion_matrix(y_test,y_pred)
cm_df=pd.DataFrame(cm,index=["Actual 0","Actual 1"],columns=["Predictd 0","Predicted 1"])
cm_df
```

```python
#Displaying results
print("Confusion Matrix:cm_df")
display(cm_df)

print(f'Accuracy Score: {acc_score}')

print("Classification Report")

print(classification_report_imbalanced(y_test, y_pred))
```

## SMOTTENN

```python
X=df.copy()
X=df.drop(columns="y")

y=df["y"]
```

```python
# Normal train-test split
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=1)
```

```python
# Use the SMOTEENN technique to perform combination sampling on the data
# Count the resampled classes
from imblearn.combine import SMOTEENN

smote_enn = SMOTEENN(random_state=0)
X_resampled, y_resampled = smote_enn.fit_resample(X, y)
Counter(y_resampled)
```

```python
# Fit a Logistic regression model using random undersampled data
from sklearn.linear_model import LogisticRegression
lgmodel = LogisticRegression(solver='lbfgs', random_state=1)
lgmodel.fit(X_resampled, y_resampled)
```

```python
# Display the confusion matrix
from sklearn.metrics import confusion_matrix

y_pred = lgmodel.predict(X_test)
cm_sm=confusion_matrix(y_test, y_pred)
cm_df=pd.DataFrame(cm_sm,index=["Actual 0","Actual 1"],columns=["Predictd 0","Predicted 1"])
cm_df
```

```python
# Calculate the Balanced Accuracy Score
from sklearn.metrics import balanced_accuracy_score
bal_acc_score=balanced_accuracy_score(y_test, y_pred)
bal_acc_score
```

```python
# Print the imbalanced classification report
from imblearn.metrics import classification_report_imbalanced

print(classification_report_imbalanced(y_test, y_pred))
```

```python
#Displaying results
print("Confusion Matrix:cm_df")
display(cm_df)

print(f'Accuracy Score: {bal_acc_score}')

print("Classification Report")

print(classification_report_imbalanced(y_test, y_pred))
```

Class imbalance refers to a situation in which the existing classes in a dataset aren't equally represented. 
It occurs when one class is much larger than the other class. 
For example, if you work for a credit card company and want to detect fraudulent transactions, you will deal with many more non-fraudulent transactions than fraudulent ones. In this case, the non-fraudulent class is much larger than the fraudulent class.\
